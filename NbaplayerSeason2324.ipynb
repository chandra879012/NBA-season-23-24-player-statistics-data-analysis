## **Team Peformance**

```python
# Calculate Net Rating
team_stats['Net_Rtg'] = team_stats['ORtg'] - team_stats['DRtg']

# Rank Teams by Net Rating
effective_teams = team_stats.sort_values(by='Net_Rtg', ascending=False)

# Display the top teams
effective_teams.head()

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 8))
sns.barplot(x=effective_teams.index, y=effective_teams['Net_Rtg'], palette='viridis')
plt.title('Team Effectiveness by Net Rating')
plt.xlabel('Team')
plt.ylabel('Net Rating')
plt.xticks(rotation=90)
plt.show()
```

```python
# Aggregating data by team with corrected columns
team_stats = df.groupby('TEAM').agg({
    'GP': 'sum',
    'MPG': 'mean',
    'USG%': 'mean',
    'TO%': 'mean',
    'FTA': 'sum',
    'FT%': 'mean',
    '2P%': 'mean',
    '3P%': 'mean',
    'PPG': 'sum',
    'RPG': 'mean',
    'APG': 'mean',
    'SPG': 'mean',
    'BPG': 'mean',
    'TPG': 'mean',
    'ORtg': 'mean',
    'DRtg': 'mean',
}).reset_index()

# Adding a few more calculated statistics for analysis
team_stats['Net Rating'] = team_stats['ORtg'] - team_stats['DRtg']
team_stats['Total P+R+A'] = team_stats['PPG'] + team_stats['RPG'] + team_stats['APG']

# Sorting teams by Net Rating and Total P+R+A to identify top teams
top_teams = team_stats.sort_values(by=['Net Rating', 'Total P+R+A'], ascending=[False, False])

top_teams.head(10)
```

```python
# Plotting Net Rating and Total P+R+A for the top 10 teams
fig, ax1 = plt.subplots(figsize=(14, 8))

# Sorting the teams by Net Rating for better visualization
top_teams_sorted = top_teams.head(10).sort_values(by='Net Rating', ascending=False)

# Plotting Net Rating
color = 'tab:blue'
ax1.set_xlabel('Team')
ax1.set_ylabel('Net Rating', color=color)
ax1.bar(top_teams_sorted['TEAM'], top_teams_sorted['Net Rating'], color=color, alpha=0.6, label='Net Rating')
ax1.tick_params(axis='y', labelcolor=color)

# Creating a second y-axis for Total P+R+A
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Total P+R+A', color=color)
ax2.plot(top_teams_sorted['TEAM'], top_teams_sorted['Total P+R+A'], color=color, marker='o', linestyle='dashed', linewidth=2, markersize=8, label='Total P+R+A')
ax2.tick_params(axis='y', labelcolor=color)

# Adding title and legend
fig.suptitle('Top 10 NBA Teams Performance Metrics', fontsize=16)
fig.legend(loc='upper right', bbox_to_anchor=(1, 0.9))

# Adding grid for better readability
ax1.grid(axis='y', linestyle='--', alpha=0.7)

plt.xticks(rotation=45)
plt.show()
```

```python
# Rank Teams by Defensive Rating (lower is better)
strong_defense_teams = team_stats.sort_values(by='DRtg')

# Display the top teams with the strongest defense
strong_defense_teams[['DRtg']].head()

# Assuming team_stats DataFrame has a 'TEAM' column with team names
plt.figure(figsize=(12, 6))
sns.barplot(x='TEAM', y='DRtg', data=team_stats.sort_values(by='DRtg', ascending=True), palette='Blues_d')
plt.title('Teams with the Strongest Defense')
plt.xlabel('Team')
plt.ylabel('Defensive Rating (DRtg)')
plt.xticks(rotation=45)  # Rotate team names for better visibility if needed
plt.show()
```

```python
# Assuming team_stats DataFrame has a 'TEAM' column with team names
plt.figure(figsize=(12, 6))
sns.barplot(x='TEAM', y='ORtg', data=team_stats.sort_values(by='ORtg', ascending=False), palette='Reds_d')
plt.title('Teams with the Strongest Offense')
plt.xlabel('Team')
plt.ylabel('Offensive Rating (ORtg)')
plt.xticks(rotation=45)  # Rotate team names for better visibility if needed
plt.show()
```

## **Analyze MVP Regular Season NBA**

```python
# Top Players by Key Metrics
top_scorers = df.sort_values(by='PPG', ascending=False).head(10)
top_rebounders = df.sort_values(by='RPG', ascending=False).head(10)
top_assist = df.sort_values(by='APG', ascending=False).head(10)

# Composite Score for MVP Potential (e.g., average rank in key metrics)
df['MVP_Score'] = (
    df['PPG'].rank(ascending=False) +
    df['RPG'].rank(ascending=False) +
    df['APG'].rank(ascending=False) +
    df['VI'].rank(ascending=False)  # VI could be a player impact metric
) / 4

# Rank Players by MVP Score
potential_mvp = df.sort_values(by='MVP_Score').head(10)

# Display the results
potential_mvp[['NAME', 'PPG', 'RPG', 'APG', 'VI', 'MVP_Score']]
```

A little explanation MVP Score, the lower the score shown, the more likely a player is to get the title of best player.

```python
# Visualization of Potential MVP Candidates

# Select the top 5 potential MVP candidates
top_mvp_candidates = potential_mvp.head(5)

# Plot the key metrics for the top 5 MVP candidates
plt.figure(figsize=(14, 8))

# PPG
plt.subplot(2, 2, 1)
sns.barplot(x='NAME', y='PPG', data=top_mvp_candidates, palette='viridis')
plt.title('Points Per Game (PPG)')
plt.xlabel('Player')
plt.ylabel('PPG')

# RPG
plt.subplot(2, 2, 2)
sns.barplot(x='NAME', y='RPG', data=top_mvp_candidates, palette='viridis')
plt.title('Rebounds Per Game (RPG)')
plt.xlabel('Player')
plt.ylabel('RPG')

# APG
plt.subplot(2, 2, 3)
sns.barplot(x='NAME', y='APG', data=top_mvp_candidates, palette='viridis')
plt.title('Assists Per Game (APG)')
plt.xlabel('Player')
plt.ylabel('APG')

# VI
plt.subplot(2, 2, 4)
sns.barplot(x='NAME', y='VI', data=top_mvp_candidates, palette='viridis')
plt.title('Player Impact (VI)')
plt.xlabel('Player')
plt.ylabel('VI')

plt.tight_layout()
plt.show()
```

```python
# Define the key metrics for analysis
key_metrics = ['PPG', 'RPG', 'APG', 'VI']

# Select the top players based on a composite score of key metrics
df['Composite_Score'] = (
    df['PPG'].rank(ascending=False) +
    df['RPG'].rank(ascending=False) +
    df['APG'].rank(ascending=False) +
    df['VI'].rank(ascending=False)
) / 4

# Identify the top players
top_players = df.sort_values(by='Composite_Score').head(10)

# Display the top players' stats
top_players_stats = top_players[['NAME', 'TEAM', 'PPG', 'RPG', 'APG', 'VI', 'USG%']]

# Visualization of top players' stats
import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size
plt.figure(figsize=(16, 10))

# Plot the key metrics for the top players
metrics_melted = pd.melt(top_players_stats, id_vars=['NAME', 'TEAM'], value_vars=key_metrics, var_name='Metric', value_name='Value')
sns.barplot(x='NAME', y='Value', hue='Metric', data=metrics_melted, palette='viridis')
plt.title('Top Players Stats Comparison')
plt.xlabel('Player')
plt.ylabel('Value')
plt.legend(title='Metric')
plt.xticks(rotation=45)
plt.show()
```
